```python
# to check target feature distribution( what type of crop can be grown)
crop['label'].value_counts()
```

```python
features = crop.columns.to_list()
features.remove('label')
print(features)
```

```python
# Visualize the distribution of each feature
import seaborn as sns
fig, ax =plt.subplots(3,3,figsize = (15,15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop[i], ax=subplot, kde=True)
    subplot.set_title(i)
plt.tight_layout
plt.show()
```

```python
#Ploting a scatter plot(bivariate analysis to know relationship b/w features)
fig, ax = plt.subplots(3,3, figsize=(15,15))
for i, subplot in zip(features, ax.flatten()):
    sns.scatterplot(x=i, y='label', data = crop,ax= subplot)
plt.tight_layout
plt.show()
```

```python
#finding outliers - boxplot
fig, ax = plt.subplots(3,3,figsize=(15,15))
for i, subplot in zip (features, ax.flatten()):
    sns.boxplot(crop[i], ax=subplot)
plt.tight_layout
plt.show()
```

```python
#correlation of features
num_cols = crop.select_dtypes(include=[np.number])
num_cols
```

```python
#correletion matrix
corr = num_cols.corr()
corr
```

```python
#vizualizing with heat map
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
